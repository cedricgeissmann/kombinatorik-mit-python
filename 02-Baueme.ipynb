# Bäume zeichnen lassen

Wir können den Computer auch verwenden um Dinge zu visualisieren. Das kann bei Bäumen sehr praktisch sein, kann aber auch sehr schnell unübersichtlich werden.

Wenn wir einen Baum zeichnen wollen, dann ist das eine andere Repräsentation von einem Zustandsraum. Genauer gesagt gibt der Baum an wie der Zustandsraum aufgebaut werden kann. Wenn wir uns als Beispiel das werfen einer Münze betrachten, und wir werfen diese Münze genau 2 mal. Dann können wir den Zustandsraum so interpretieren, das an der Stelle 1 der erste Wurf der Münze steht, und an der Stelle 2 der zweite Wurf. Für mehr Würfe geht es dann immer so weiter.

In der nächsten Zelle finden Sie Code der das Werfen einer Münze 2 mal nacheinander simuliert.

```python
from itertools import product

muenze = ["K", "Z"]  # wir definieren die verschiedenen Ergebnisse eines Münzwurfs
Z = product(muenze, repeat=2)  # Wir erzeugen alle Möglichkeiten wenn diese Münze 2 mal geworfen wird.
Z = list(Z)  # muss in eine Liste umgewandelt werden, damit wir weiter arbeiten können.

display(Z)  # Gibt den Zustandsraum für den Münzwurf aus.
```

In der Zelle oben haben wir uns systematisch alle Ergebnisse von einem doppelten Münzwurf erzeugen lassen, also den Zustandsraum für dieses Experiment erstellt. Betrachten wir nun die einzelnen Ergebnisse, sehen wir dass das Element ganz links in der `()`-Klammer, immer den Wurf der ersten Münze repäsentiert, und das Element rechts dann den Wurf der zweiten Münze. Mit dieser Interpretation kann auch ein Baum aufgebaut werden.

Der Code in der nächsten Zelle zeichnet Ihnen einen Baum von dem Zustandsraum. Dazu müssen Sie nur die Helferfunktionen importieren.

```python
from helpers import construct_tree  # Helferfunktion um einen Baum zu zeichnen importieren

display(construct_tree(Z))  # Den Baum aus einem Zustandsraum Z erstellen und anzeigen.
```

## Grössere Bäume zeichnen

Wir können auch grössere Bäume zeichnen, wobei dann schnell mal der Platz ausgeht und das ganze unübersichtlich wird. In der nächsten Zelle zeichnen wir einen Baum von dem gleichen Experiment wenn die Münze mehrfach geworfen wird.

```python
from itertools import product
from helpers import construct_tree

muenze = ["K", "Z"] 
Z = list(product(muenze, repeat=4))

display(Z)
display(construct_tree(Z))
```

## Teile des Zustandsraum als Baum zeichnen

Wir können auch Ereignisse als Bäume zeichnen. Dazu müssen wir den Zustandsraum filtern, so das wir nur noch die gewünschten Ergebisse im Ereignis haben.

In der nächsten Zelle erstellen wir einen Filter der uns das Ereignis mindestens 2 mal Kopf zu werfen ergibt.

```python
def min_2_heads(elem):
    if elem.count('K') >= 2:
        return True
    return False

res = list(filter(min_2_heads, Z))
display(construct_tree(res))
```
